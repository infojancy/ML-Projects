```python
data_A.rename(columns={'CTSA':'CTSA_ap',
                      'GOVT':'GOVT_ap',
                      'GOVT AIDED': 'GOVT_AIDED_ap',
                      'INDEPENDENT':'INDEPENDENT_ap',
                      'JNV':'JNV_ap', 
                      'KV':'KV_ap'}, inplace=True)
```

```python
data_A.reset_index(drop=True)
data_R.reset_index(drop=True)
```

```python
data_merged = data_R.merge(data_A, on='Region')
```

```python
data_merged.drop(columns=['Status_x', 'Status_y', 'reg_ctsa'], inplace=True)
```

```python
data_merged.rename(columns={'GOVT AIDED':'GOVT_AIDED'},inplace=True)
```

```python
data_merged['GOVT_%'] = (data_merged.GOVT_ap *100 )/ data_merged.GOVT
```

```python
data_merged
```

```python
data_merged['GOVT_AID_%'] = (data_merged.GOVT_AIDED_ap *100 )/ data_merged.GOVT_AIDED
```

```python
data_merged['INDEP_%'] = (data_merged.INDEPENDENT_ap *100 )/ data_merged.INDEPENDENT
```

```python
data_merged['CTSA_%'] = (data_merged.CTSA_ap *100 )/ data_merged.CTSA
```

```python
data_merged['JNV_%'] = (data_merged.JNV_ap *100 )/ data_merged.JNV
```

```python
data_merged['KV_%'] = (data_merged.KV_ap *100 )/ data_merged.KV
```

```python
data_merged
```

```python
data_merged.rename(columns={'GOVT_%':'GOVT_perc',
                      'GOVT_AID_%':'GOVT_AID_perc',
                      'CTSA_%': 'CTSA_perc',
                      'INDEP_%':'INDEPENDENT_perc',
                      'JNV_%':'JNV_perc', 
                      'KV_%':'KV_perc'}, inplace=True)
```
